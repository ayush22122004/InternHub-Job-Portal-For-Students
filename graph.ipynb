```python
import yfinance as yf
import matplotlib.pyplot as plt


def scrape_stock_performance(symbol):
    try:
        # Fetch historical stock data using yfinance
        stock = yf.Ticker(symbol)
        historical_data = stock.history(period="5y")

        # Check if historical data is empty
        if not historical_data.empty:
            # Extract years and revenue from the data
            years = historical_data.index.year
            revenue = historical_data['Close'] * historical_data['Volume'] / 10**6  # Adjust column names if needed
            return years, revenue
        else:
            print(f"\nNo historical data available for {symbol}")
            return None, None
    except Exception as e:
        print(f"Failed to retrieve data for {symbol}: {e}")
        return None, None

def plot_performance(symbol, years, revenue):
    # Plot the company's performance over the past years
    plt.plot(years, revenue, label='Revenue (in millions)')
    
    plt.xlabel('Year')
    plt.ylabel('Amount (in millions)')
    plt.title(f'{symbol} Performance Over Past 5 Years')
    plt.legend()
    
    # Save the plot as an image or display it
    plt.savefig(f'{symbol}_performance.png')
    plt.show()

if __name__ == "__main__":
    companies = {
        'AAPL': 'Apple Inc.',
        'MSFT': 'Microsoft Corporation',
        'AMZN': 'Amazon.com, Inc.',
        'GOOGL': 'Alphabet Inc.',
        'FB': 'Facebook, Inc.',
        'TSLA': 'Tesla, Inc.',
        'NFLX': 'Netflix, Inc.',
        'JNJ': 'Johnson & Johnson',
        'KO': 'The Coca-Cola Company',
        'PFE': 'Pfizer Inc.'
    }
    
    while True:
        for symbol, name in companies.items():
            print(symbol, name)

        symbol = input("Enter a company symbol or type 'exit' to quit: ").upper()
        if symbol == 'EXIT':
            break
        elif symbol in companies:
            name = companies[symbol]
            years, revenue = scrape_stock_performance(symbol)
            if years is not None and revenue is not None:  
                plot_performance(name, years, revenue)
        else:
            print("Company symbol not found. Please enter a valid symbol.")
```
